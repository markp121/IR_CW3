```python
import pandas as pd
import numpy as np
from ast import literal_eval
import re

games = pd.read_csv('IR_CW3/onlygames_steam_appdata_extracleaned.csv')
games.developers = games.developers.replace(np.NaN, '[]')
```

```python
games.loc[:,'publishers'] = games.loc[:,'publishers'].apply(lambda x: literal_eval(x))
games.loc[:,'developers'] = games.loc[:,'developers'].apply(lambda x: literal_eval(x))
```

```python
games.release_date = games.release_date.replace('.*(?=(\d{1,2} [A-Za-z]+, \d{4}))', "", regex=True)\
                                       .replace('(?<=\d{4}).*', "", regex=True)\
                                       .replace('.*\'(?=([A-Za-z]+ \d{1,2}, \d{4}))', "", regex=True)\
                                       .replace(r'([A-Za-z]+) (\d{1,2})', r'\2 \1', regex=True)\
                                       .replace(r"({'coming_soon': True, 'date':).*", 'ur', regex=True)\
                                       .replace(r"({'coming_soon': False, 'date':).*", np.NaN, regex=True)\
                                       .replace(r",", "", regex=True)\
                                       .replace('(?<=([A-Za-z]{3}))[a-z]', "", regex=True)\
                                       .replace('ur', 'coming_soon')



games.about_the_game = games.about_the_game.replace('<.*?>', '', regex=True)\
                                           .replace(r'(\t)+', '', regex=True)\
                                           .replace(r'(\r\n)+', '', regex=True)\
                                           .replace(r'\n+', '', regex=True)\
                                           .replace('\s{3,}', " ", regex=True)


games = games.dropna()

games.genres = games.genres.apply(lambda x: re.findall("[A-Z][\w\s]+", x))
games.categories = games.categories.apply(lambda x: re.findall("[A-Z][\w\s/\-]+", x))
```

```python
pd.to_pickle(games, 'cleaned_steam_data_V2')
```
